```
#download the stopwords package
nltk.download('stopwords')
```

```
def process_text(text):
  
  #removing punctuations
  nopunc = [char for char in text if char not in string.punctuation]
  nopunc = ''.join(nopunc)
  
  #removing stopwords(i.e. a, the, an, in etc)
  clean_words = [word for word in nopunc.split() if word.lower() not in stopwords.words('english')]

  #return a list of clean words
  return clean_words
```

```
#showing the list of tokens aka lemmas
df['text'].head().apply(process_text)
```

```
#convert a collection of text into a matrix of tokens
from sklearn.feature_extraction.text import CountVectorizer
messages_bow = CountVectorizer(analyzer=process_text).fit_transform(df['text']) 
print(messages_bow)

#---Subject--there--Always--
#0--1--------1------0------
#1--1--------1------1------
```

Splitting the data into train and test

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(messages_bow, df['label_num'], test_size=0.2) #label_num is our output
```

```
messages_bow.shape
```

```
#training the model
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB().fit(X_train, y_train)
```

Evaluating the model

```
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
pred = classifier.predict(X_train)
print(classification_report(y_train, pred))
print()
print('Confusion martix: \n', confusion_matrix(y_train, pred))
print()
print('Accuracy: ', accuracy_score(y_train, pred))
```

```
print(classifier.predict(X_test))
print(y_test.values)
```

```
#evaluating our accuracy with the test dataset
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
pred = classifier.predict(X_test)
print(classification_report(y_test, pred))
print()
print('Confusion martix: \n', confusion_matrix(y_test, pred))
print()
print('Accuracy: ', accuracy_score(y_test, pred))
```
